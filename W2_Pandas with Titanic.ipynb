## Add and Modify Columns

```python
train["Category"] = "Titanic"
train.head()
```

```python
train.shape
```

```python
train["Ids"] = range(891) #Only values that fit the 'size' can fit in.
train.head()
```

```python
train["FamilySize"] = train["SibSp"] + train["Parch"] + 1
train[["SibSp", "Parch", "FamilySize"]].head()
```

```python
train["Nationality_FR"] = train["Embarked"] == "C"
train["Nationality_UK"] = train["Embarked"] != "C"

train[["Embarked", "Nationality_FR", "Nationality_UK"]].head()
```

```python
train.loc[train["Embarked"] == "C", "Nationality"] = "France"
train.loc[train["Embarked"] != "C", "Nationality"] = "England"

train[["Embarked", "Nationality"]].head()
```

```python
train["LowFare"] = train["Fare"] < 15
train["MediumFare"] = (train["Fare"] >= 15) & (train["Fare"] < 35)
train["HighFare"] = train["Fare"] >= 35

train.loc[train["Fare"] < 15, "Faretype"] = "LowFare"
train.loc[(train["Fare"] >= 15) & (train["Fare"] < 35), "Faretype"] = "MediumFare"
train.loc[train["Fare"] >= 35, "Faretype"] = "HighFare"

train[["Fare", "Faretype", "LowFare", "MediumFare", "HighFare"]].head(10)
```

```python
mean_age = train["Age"].mean()
train.loc[train["Age"].isnull(), "Age"] = mean_age
train.head()
```

```python
train.to_csv("train_modified.csv")
```
