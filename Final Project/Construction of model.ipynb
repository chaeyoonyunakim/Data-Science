```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
df = pd.read_csv('master_merged.csv')
df
```

```python
y = df.loc[:, "suicides_per_100k"]
```

```python
nominal_cols = df.dtypes[df.dtypes == object].reset_index().iloc[:, 0]
X = df.drop(columns = nominal_cols, axis = 1)
X.drop(columns = ["suicides_per_100k"], axis = 1, inplace = True)
X
```

```python
lm = LinearRegression()
lm.fit(X, y.values)
```

```python
lm.coef_
```

```python
mean_absolute_error(lm.predict(X), y.values)
```
